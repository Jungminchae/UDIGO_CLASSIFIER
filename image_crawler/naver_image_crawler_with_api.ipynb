```python
import os
import json
import requests
import urllib.request
from pathlib import Path

from credentials import naver
```

```python
client_id = naver['client_id']
client_secret = naver['client_secret']
display = 100
page = 11
savePath = './images/naver_api'
```

```python
def crawl_image(keyword):
    # 이미지 저장할 폴더 폴더 생성
    image_path = os.path.join(savePath, keyword)
    if not os.path.exists(image_path):
        os.makedirs(image_path, exist_ok=True)
    
    # 아스키코드 형식이 아닌 글자 URL 인코딩
    encText = urllib.parse.quote(keyword)

    file_n = 1
    for i in range(1, page):
        print('check i : ', i)
        
        # display : 검색 결과 출력 건수, start : 검색 시작 위치
        url = f"https://openapi.naver.com/v1/search/image?query={encText}&display={display}&start={i*100}"
        headers = {
            'X-Naver-Client-Id': client_id,
            'X-Naver-Client-Secret': client_secret
        }
        req = requests.get(url, headers=headers)

        if req.status_code == 200:
            html = req.text
            html = json.loads(html)['items']
            for img in html:
                try:
                    # 이미지링크 확인
                    img_url = img['link']
                    print(img_url)

                    # 저장 파일명 및 경로
                    file_name = Path(img_url).name                  # Path().name : 디렉토리가 제외된 파일 이름
                    FileName = os.path.join(image_path, file_name)  # 이미지를 저장할 경로와 file_name을 합침

                    # 파일명 출력
                    print('full name : {}'.format(FileName))

                    # 이미지 다운로드 URL 요청
                    urllib.request.urlretrieve(img_url, FileName)
                    file_n += 1

                except urllib.error.HTTPError as e:  # 서버가 요청을 이행할 수 없을 때 - 리디렉션, 401, 403, 404 등 
                    print('oh no... : ', e.reason)
                    pass

                except urllib.error.URLError as e:  # 네트워크 연결이 없거나 (지정된 서버로의 경로가 없거나), 지정된 서버가 없을 때
                    print('oh no... : ', e.reason)
                    pass

            # 다운로드 완료 시 출력
            print("--------download succeeded--------")

        else:
            print("Error Code:" + str(req.status_code))
```

```python
crawl_image('세빛섬')
```
