```python
import os
import json
import requests
import datetime
import urllib.request

from pathlib import Path
from credentials import kakao
```

```python
savePath = './images/kakao_api'
Authorization = kakao['Authorization']
```

```python
def crawl_image(keyword):
    # 이미지 저장할 폴더 폴더 생성
    image_path = os.path.join(savePath, keyword)
    if not os.path.exists(image_path):
        os.makedirs(image_path, exist_ok=True)

    # kakao는 페이지 요청을 1부터 50까지 가능
    # 참고 : https://developers.kakao.com/docs/latest/ko/daum-search/dev-guide#search-image
    for page in range(1, 51):
        url = f"https://dapi.kakao.com/v2/search/image?query={keyword}&page={page}"
        headers = {'Authorization': Authorization}
        req = requests.get(url, headers=headers)  # requests에 get요청하여 response를 req 변수에 담음

        # status code가 200(정상)이라면
        if req.status_code == 200:
            html = req.text
            html = json.loads(html)['documents']
            
            for el in html:
                try:
                    # 이미지링크 확인
                    url = el['image_url']
                    print(url)

                    # 저장 파일명 및 경로 
                    file_name = Path(url).name  # Path().name : 디렉토리가 제외된 파일 이름
                    FileName = os.path.join(image_path, file_name)  # 이미지를 저장할 경로와 file_name을 합침

                    # 파일명 출력
                    print('full name : {}'.format(FileName))

                    # 이미지 다운로드 URL 요청
                    urllib.request.urlretrieve(url, FileName)

                except urllib.error.HTTPError as e:  # 서버가 요청을 이행할 수 없을 때 - 리디렉션, 401, 403, 404 등 
                    print('oh no... : ', e.reason)
                    pass

                except urllib.error.URLError as e:  # 네트워크 연결이 없거나 (지정된 서버로의 경로가 없거나), 지정된 서버가 없을 때
                    print('oh no... : ', e.reason)
                    pass
                except OSError as e:  # 경로 혹은 파일 이름에 사용할 수 없는 문자가 들어있거나 파일 이름이 너무 긴 경우
                    print('oh no... : ', e)
                    # 파일명을 현재 시간으로 변경
                    now = datetime.datetime.now()
                    nowDatetime = now.strftime('%Y%m%d_%H%M%S')
                    FileName = os.path.join(image_path, nowDatetime)
                    urllib.request.urlretrieve(url, FileName)
                    pass
    
            # 다운로드 완료 시 출력
            print("--------download succeeded--------")

        else:
            print("Error Code:" + str(req.status_code))
```

```python
crawl_image('세빛섬')
```
